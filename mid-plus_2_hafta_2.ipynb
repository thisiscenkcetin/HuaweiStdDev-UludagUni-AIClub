**1. CNN‚Äôin Temel Amacƒ±**

CNN‚Äôler, √∂zellikle g√∂r√ºnt√º verilerini i≈ülerken, √∂nemli olan √∂znitelikleri (features) otomatik olarak √∂ƒürenir. √ñrneƒüin bir araba resmi i√ßin:

- Kenarlar (edge)

- K√∂≈üeler (corner)

- Doku (texture)

- Y√ºksek seviyeli yapƒ±lar (tekerlek, far vs.)

- Bu √∂znitelikler, CNN‚Äôin feature map dediƒüimiz ara katmanlarƒ±nda temsil edilir.

**2. Konvol√ºsyon Katmanƒ± (Convolution Layer)**

CNN‚Äôin kalbi konvol√ºsyon katmanlarƒ±dƒ±r.

- Girdi: √ñrneƒüin bir  ùêª√óùëä√óùê∂ boyutunda bir resim (y√ºkseklik, geni≈ülik, kanal).

- Filtre (kernel): K√º√ß√ºk bir matris (√∂rn. 3x3 veya 5x5)

- ƒ∞≈ülem: Filtre, g√∂r√ºnt√º √ºzerinde kaydƒ±rƒ±lƒ±r (sliding), her konumda bir √ßarpma ve toplama i≈ülemi yapƒ±lƒ±r.

Matematiksel olarak:

$$(feature_map)_{i,j}=\sum\sum_nI_{i+m,j+n}\cdot K_{m,n}$$

- $I:$ Girdi g√∂r√ºnt√º

- $K:$ Filtre (Kernel)

- $i,j:$ Feature map √ºzerindeki konum

**Ama√ß:** G√∂r√ºnt√ºdeki belirli desenleri (kenar, k√∂≈üe, √ßizgi) yakalamak.

**Filtreler**

Dikey (sol-saƒü deƒüi≈üim):
$$K_{vert} = \begin{bmatrix} -1 & 0 & 1 \\ -1 & 0 & 1 \\ -1 & 0 & 1 \end{bmatrix}$$
Yatay (√ºst-alt deƒüi≈üim):
$$
K_{hor} = \begin{bmatrix} -1 & -1 & -1 \\ 0 & 0 & 0 \\ 1 & 1 & 1 \end{bmatrix}$$

3. Aktivasyon Fonksiyonu

Konvol√ºsyon i≈üleminden sonra ReLU gibi bir aktivasyon uygulanƒ±r:

$$f(x)=\max(0,x)$$

- Negatif deƒüerler sƒ±fƒ±rlanƒ±r, pozitif deƒüerler korunur.

- Bu sayede non-linearity eklenir ve daha karma≈üƒ±k desenler √∂ƒürenilebilir.

**4. Pooling Katmanƒ± (√ñzellik Se√ßme / Azaltma)**

- CNN‚Äôlerde genellikle max pooling veya average pooling kullanƒ±lƒ±r.

- Ama√ß: Feature map‚Äôin boyutunu k√º√ß√ºltmek, √∂nemli bilgileri koruyarak gereksiz detaylarƒ± azaltmak.

√ñrnek: 2x2 max pooling ‚Üí 4 pikselin maksimum deƒüeri alƒ±nƒ±r, boyut yarƒ±ya iner.

**5. Derin Katmanlarda Y√ºksek Seviyeli √ñzellikler**

- ƒ∞lk katmanlar ‚Üí Kenarlar, √ßizgiler, basit dokular

- Orta katmanlar ‚Üí Basit ≈üekiller, k√∂≈üeler, motifler

- Derin katmanlar ‚Üí Kompleks nesne par√ßalarƒ± (tekerlek, g√∂z, araba farƒ±)

Yani CNN, otomatik olarak hiyerar≈üik √∂zellikler √ßƒ±karƒ±r.

**6. Feature Map‚Äôin √ñnemi**

- Her konvol√ºsyon + aktivasyon katmanƒ± ‚Üí feature map √ºretir.

- Feature map, hangi piksel b√∂lgelerinin hangi √∂zellikleri ta≈üƒ±dƒ±ƒüƒ±nƒ± g√∂sterir.

Daha derin katmanlarda feature map, orijinal resmin k√º√ß√ºk bir ‚Äútemsilini‚Äù ta≈üƒ±r, nesnenin y√ºksek seviyeli √∂zelliklerini g√∂sterir.

**√ñzet**

CNN‚Äôde √∂zellik √ßƒ±karƒ±mƒ± s√ºreci ≈ü√∂yle √∂zetlenebilir:

- Konvol√ºsyon: Kenar ve temel desenleri yakalar

- ReLU: Negatif deƒüerleri temizler, non-linearity ekler

- Pooling: √ñnemli √∂zellikleri se√ßer, boyutu k√º√ß√ºlt√ºr

- Derinle≈üme: Daha karma≈üƒ±k ve y√ºksek seviyeli √∂zellikleri √∂ƒürenir

Sonu√ßta, bu √ßƒ±karƒ±lan √∂zellikler sƒ±nƒ±flandƒ±rma veya nesne tespiti gibi g√∂revlerde kullanƒ±lƒ±r.

CNN Mimarisini anlamak i√ßin doƒürudan bir √∂rneƒüi manuel olarak incelemek ve bazƒ± kavramlarƒ± √∂rnek √ºzerinde g√∂rmek daha iyi olacaktƒ±r

*Elimizde 100x100 tipinde matris var ve her bir elemanƒ± her pikseldeki renk skalasƒ±nƒ± temsil ediyor siyah = 0 ve beyaz = 255, tam bu noktada kenarƒ± belirleyebilmek i√ßin kartezyen sistem √ºzerinde d√º≈ü√ºnd√ºƒü√ºm√ºzde sayƒ±nƒ±n a≈üƒ±rƒ± farkla deƒüi≈ümesi bize bir deƒüi≈üimin olduƒüunu sezdirebilir ve bilgisayara bu ≈üekilde aktarmakta zorlanabiliriz tam bu noktada kernel metodunu kullanƒ±rƒ±z*

A≈üaƒüƒ±daki filtre (kernel) matrisi ile **Dikey Kenar'larƒ±** tespit edebiliriz

$$\begin{bmatrix}-1&0&1\\-1&0&1\\-1&0&1\end{bmatrix}$$

√ñrneƒüin:

$$\begin{bmatrix}0&0&0\\0&0&0\\0&0&0\end{bmatrix}$$

**Resmimizde tam bu koordinatlardaki b√∂lgeyi inceleyelim hepsi sƒ±fƒ±rlardan 3x3 piksel boyutundan olu≈üan herhangi siyah renkte bir yer**

$$C=\sum_{i,j}K_{i,j}\cdot R_{i,j}$$

Form√ºl√ºn√º de vermi≈ü olalƒ±m, basit√ße a≈üaƒüƒ±daki element-wise i≈ülemini yaparak

$$(‚àí1‚ãÖ0)+(0‚ãÖ0)+(1‚ãÖ0)=0$$
$$(‚àí1‚ãÖ0)+(0‚ãÖ0)+(1‚ãÖ0)=0$$
$$(‚àí1‚ãÖ0)+(0‚ãÖ0)+(1‚ãÖ0)=0$$
$$0+0+0 = 0$$

Burada hi√ßbir ≈üey yok, sonu√ß 0 sezgisini elde edebiliriz.

**Not:** Element-wise i≈ülemi 3x3 tipinde iki matrisin eleman elemana √ßarpƒ±mƒ±dƒ±r normal matris √ßarpƒ±mƒ± i≈üleminden farklƒ±dƒ±r.

**Beyaz karenin tam ortasƒ±nda 3x3 tipinde filtre gezerken** yine aynƒ± i≈ülemi tekrarlayarak farklƒ± bi durumla kar≈üƒ±la≈ümayacaƒüƒ±mƒ±zƒ± i≈ülem yaparak g√∂rebilirsiniz.

**A≈üaƒüƒ±da verilen matris i≈üleminde 3. s√ºtundaki beyaz deƒüerlerin √ºst √ºste sƒ±ralandƒ±ƒüƒ±nda dikey bir kenar olu≈üturduƒüunu anlayabiliriz

$$R\odot K=\begin{bmatrix}0\cdot(-1)&0\cdot0&255\cdot1\\0\cdot(-1)&0\cdot0&255\cdot1\\0\cdot(-1)&0\cdot0&255\cdot1\end{bmatrix}=\begin{bmatrix}0&0&255\\0&0&255\\0&0&255\end{bmatrix}$$

Matris elemanlarƒ±nƒ± topladƒ±ƒüƒ±mƒ±zda 

$$C=0+0+255+0+0+255+0+0+255=765$$

- Sonu√ß b√ºy√ºk bir deƒüer ‚Üí burada dikey bir deƒüi≈üim var (siyah ‚Üí beyaz).

- Eƒüer matris t√ºm√ºyle siyah ya da beyaz olsaydƒ±, toplam 0 olurdu ‚Üí deƒüi≈üim yok.

$$\begin{bmatrix}255&255&255\\255&255&255\\255&255&255\end{bmatrix}$$

**Dikey kenar ifade eden matrise yatay kenar kernelini uygularsak**

$$R\odot K_{har}=\begin{bmatrix}0\cdot(-1)&0\cdot(-1)&255\cdot(-1)\\0\cdot0&0\cdot0&255\cdot0\\0\cdot1&0\cdot1&255\cdot1\end{bmatrix}=\begin{bmatrix}0&0&-255\\0&0&0\\0&0&255\end{bmatrix}$$

$$C_{hor‚Äã}=0+0+(‚àí255)+0+0+0+0+0+255=0$$

- Sonu√ß **0** $\rightarrow$ **yatay kenar yok**

**Renk deƒüi≈üimi $\rightarrow$ Filre √ßarpƒ±mƒ±**

- Sol taraf siyah kƒ±sƒ±m, saƒü taraf beyaz kƒ±sƒ±m ise (0 $\rightarrow$ 255):

$$\begin{bmatrix}0&255&255\\0&255&255\\0&255&255\end{bmatrix}$$

$$\mathrm{toplam}=(-1\cdot0)+(0\cdot0)+(1\cdot255)=255>0$$

- Pozitif deƒüer $\rightarrow$ siyahtan beyaza **soldan saƒüa ge√ßi≈ü**

$$\mathrm{toplam}=(-1\cdot255)+(0\cdot0)+(1\cdot0)=255<0$$

- Negatif deƒüer $\rightarrow$ beyazdan siyaha ge√ßi≈ü **soldan saƒüa ge√ßi≈ü**

**√ñzet**

1. Filtreyi resim √ºzerinde kaydƒ±rƒ±yoruz (convolution)

2. Her 3x3 pencere i√ßin element-wise √ßarp ve topla

3. Ortadaki piksele sonucu yaz

4. Pozitif / negatif deƒüerler ‚Üí kenar y√∂n√ºn√º g√∂sterir

5. Ortaya √ßƒ±kan matris ‚Üí kenar haritasƒ±

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageDraw

def manuel_evrisim(resim, filtre):
    h, w = resim.shape
    f_h, f_w = filtre.shape
    # √áƒ±ktƒ± boyutu
    cikti_h = h - f_h + 1
    cikti_w = w - f_w + 1
    feature_map = np.zeros((cikti_h, cikti_w))
    
    # --- Sliding Window (Evri≈üim D√∂ng√ºs√º) ---
    for i in range(cikti_h):
        for j in range(cikti_w):
            resim_parcasi = resim[i:i+f_h, j:j+f_w]
            deger = np.sum(resim_parcasi * filtre)
            feature_map[i, j] = deger
    return feature_map

# --- Filtreleri Tanƒ±mla ---
dikey_filtre = np.array([
    [-1, 0, 1],
    [-1, 0, 1],
    [-1, 0, 1]
])

yatay_filtre = np.array([
    [-1, -1, -1],
    [ 0,  0,  0],
    [ 1,  1,  1]
])

# --- Evri≈üim Uygula ---
dikey_kenarlar = manuel_evrisim(orijinal_resim, dikey_filtre)
yatay_kenarlar = manuel_evrisim(orijinal_resim, yatay_filtre)

# --- G√∂rselle≈ütir ---
fig, ax = plt.subplots(1, 3, figsize=(15, 5))
ax[0].imshow(orijinal_resim, cmap='gray')
ax[0].set_title("Orijinal")
ax[1].imshow(dikey_kenarlar, cmap='gray')
ax[1].set_title("Dikey Kenarlar")
ax[2].imshow(yatay_kenarlar, cmap='gray')
ax[2].set_title("Yatay Kenarlar")
plt.show()
```

*Biz √∂nce bir C deƒüeri (convolution sonucu) elde ediyoruz:*

- C pozitif olabilir

- C negatif olabilir

- C k√º√ß√ºk olabilir

- C √ßok b√ºy√ºk olabilir

- Bu C deƒüeri kenarƒ±n g√ºc√ºn√º temsil ediyor.

*Ama biz kenarƒ±n y√∂n√ºyle (pozitif/negatif) uƒüra≈ümak istemiyoruz ‚Üí sadece kenar var mƒ± yok mu istiyoruz.*

*O y√ºzden √∂nce:*

- **enerji = abs(C)**

**Bu ne demek?**

- enerji sadece ger√ßekten kenar olan yerlerde b√ºy√ºkt√ºr

**Thresholding Neden Lazƒ±m?**

√á√ºnk√º convolution her yerde bir sayƒ± √ºretir.

Ama biz:

- g√ºr√ºlt√ºy√º silmek

- sadece g√º√ßl√º kenarlarƒ± bƒ±rakmak

- objelerin sƒ±nƒ±rƒ±nƒ± bulmak

isteriz.

Thresholding olmadan resim ≈ü√∂yle g√∂r√ºn√ºr:

- Her yerde bir ≈üey var

- Zayƒ±f kenarlar da e≈üit √∂nemde

- Obje tespiti patlar

Thresholding sonrasƒ±:

- Sadece ger√ßek kenarlar kalƒ±r

- Obje contour‚Äôu belirir

- Min‚Äìmax kutu i≈ülemi saƒülam √ßalƒ±≈üƒ±r

```python
def obje_bul(kenar_haritasi_1, kenar_haritasi_2):
    # 1. Kenar haritalarƒ±nƒ± birle≈ütir
    toplam_enerji = np.abs(kenar_haritasi_1) + np.abs(kenar_haritasi_2)
    
    # 2. E≈üik Deƒüer Uygula
    esik = 500
    aktif_pikseller = np.argwhere(toplam_enerji > esik)
    
    if len(aktif_pikseller) == 0:
        return "Obje Bulunamadƒ±", None
    
    # 3. Min/Max koordinatlarƒ± bul
    y_min, x_min = aktif_pikseller.min(axis=0)
    y_max, x_max = aktif_pikseller.max(axis=0)
    
    return toplam_enerji, (x_min, y_min, x_max, y_max)

# --- Obje tespiti ---
enerji_haritasi, kutu = obje_bul(dikey_kenarlar, yatay_kenarlar)

# --- Sonucu orijinal resim √ºzerine √ßiz ---
sonuc_resmi = Image.fromarray(orijinal_resim.astype('uint8')).convert('RGB')
draw = ImageDraw.Draw(sonuc_resmi)

if kutu:
    x1, y1, x2, y2 = kutu
    draw.rectangle([x1, y1, x2, y2], outline="red", width=2)
    print(f"Obje algƒ±landƒ±! Koordinatlar: {kutu}")

plt.figure(figsize=(6,6))
plt.imshow(sonuc_resmi)
plt.title("Algƒ±lanan Obje (Kƒ±rmƒ±zƒ± Kutu)")
plt.show()
```

$Piksel=(R=255,G=0,B=0)$ yani bir piksel tamamiyle kƒ±rmƒ±zƒ± olsun ye≈üil ve mavi yoƒüunluƒüu yok

≈ûimdi bir pikseli tek satƒ±rlƒ± matrisle ifade ettik kernel'i resim √ºzerinde gezdirmek i√ßin her birini ayrƒ± ayrƒ± kanalda uygulamalƒ±yƒ±z

√ñrneƒüin :

R kanalƒ±:

$$\begin{bmatrix}10&10&200\\10&10&200\\10&10&200\end{bmatrix}$$

G kanalƒ±:

$$\begin{bmatrix}0&0&50\\0&0&50\\0&0&50\end{bmatrix}$$

B kanalƒ±:

$$\begin{bmatrix}0&0&50\\0&0&50\\0&0&50\end{bmatrix}$$

3 farklƒ± kanal aynƒ± 3x3'l√ºk alandaki pikselleri ifade eder.

- (10,0,0) -> Bir piksel

- (200,50,50) -> Bir piksel

Her matriste satƒ±r ve s√ºtunlarƒ± aynƒ± olan √º√ß farklƒ± kanal tek bir pikseli ifade ederek o pikselin RGB renklerini temsil eder ve her kanala ayrƒ± ayrƒ± kernel uygulanƒ±r.

- $R_out = R * K$

- $G_out = G * K$

- $B_out = B * K$

Kernel i≈ülemini yapalƒ±m

$$(-1)*10  + 0*10  + 1*200$$

$$(-1)*10  + 0*10  + 1*200$$

$$(-1)*10  + 0*10  + 1*200$$

$$(- 10 - 10 - 10) + (200 + 200 + 200) = 570$$

570 √ßok b√ºy√ºk $\rightarrow$ **kenar vardƒ±r**

Sonu√ß: R kanalƒ±nda kenar var B kanalƒ±nda yok Bu tamamen normaldir. √á√ºnk√º ger√ßek kenar kƒ±rmƒ±zƒ± renkten kaynaklanmƒ±≈ütƒ±r. CNN bunu renk ayrƒ±≈ütƒ±rmasƒ± sayesinde daha net g√∂r√ºr

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageDraw

# --- 1. √ñNCEKƒ∞ FONKSƒ∞YONUMUZ (EVRƒ∞≈ûƒ∞M) ---
def manuel_evrisim(resim_katmani, filtre):
    h, w = resim_katmani.shape
    f_h, f_w = filtre.shape
    cikti = np.zeros((h - f_h + 1, w - f_w + 1))
    
    for i in range(cikti.shape[0]):
        for j in range(cikti.shape[1]):
            parca = resim_katmani[i:i+f_h, j:j+f_w]
            cikti[i, j] = np.sum(parca * filtre)
    return cikti

# --- 2. RENKLƒ∞ RESƒ∞M OLU≈ûTURMA ---
def renkli_resim_olustur():
    # RGB Modunda (Renkli) Siyah Zemin
    img = Image.new('RGB', (100, 100), color=(0, 0, 0)) 
    draw = ImageDraw.Draw(img)
    
    # 1. KIRMIZI KARE (Sadece R kanalƒ± dolu)
    # R=255, G=0, B=0
    draw.rectangle([10, 30, 40, 70], fill=(255, 0, 0))
    
    # 2. TURKUAZ KARE (Ye≈üil + Mavi kanallarƒ± dolu)
    # R=0, G=255, B=255
    draw.rectangle([60, 30, 90, 70], fill=(0, 255, 255))
    
    return np.array(img)

# Resmi olu≈ütur ve kanallara ayƒ±r (Slicing)
renkli_resim = renkli_resim_olustur()
r_kanali = renkli_resim[:, :, 0] # Kƒ±rmƒ±zƒ± Katman
g_kanali = renkli_resim[:, :, 1] # Ye≈üil Katman
b_kanali = renkli_resim[:, :, 2] # Mavi Katman

# --- 3. Fƒ∞LTRELEME ƒ∞≈ûLEMƒ∞ ---
# Dikey Kenar Filtresi
filtre = np.array([
    [-1, 0, 1],
    [-1, 0, 1],
    [-1, 0, 1]
])

# Filtreyi her kanala ayrƒ± ayrƒ± uygula (3 Boyutlu i≈ülem sim√ºlasyonu)
r_cikti = manuel_evrisim(r_kanali, filtre)
g_cikti = manuel_evrisim(g_kanali, filtre)
b_cikti = manuel_evrisim(b_kanali, filtre)

# --- 4. SONU√áLARI Bƒ∞RLE≈ûTƒ∞RME (TOPLAMA) ---
# Nihai Feature Map = R_sonuc + G_sonuc + B_sonuc
# Kenarlarƒ±n g√ºc√ºn√º g√∂rmek i√ßin mutlak deƒüerlerini toplayalƒ±m
nihai_harita = np.abs(r_cikti) + np.abs(g_cikti) + np.abs(b_cikti)

# --- 5. G√ñRSELLE≈ûTƒ∞RME ---
fig, ax = plt.subplots(2, 3, figsize=(15, 10))

# √úst Sƒ±ra: Orijinal Resim ve Kanallar
ax[0, 0].imshow(renkli_resim)
ax[0, 0].set_title("Orijinal Renkli Resim")
ax[0, 1].imshow(r_kanali, cmap='Reds')
ax[0, 1].set_title("Kƒ±rmƒ±zƒ± Kanalƒ± (Sadece Soldaki Var)")
ax[0, 2].imshow(g_kanali, cmap='Greens') # Turkuazƒ±n i√ßinde ye≈üil olduƒüu i√ßin saƒüdaki g√∂r√ºn√ºr
ax[0, 2].set_title("Ye≈üil Kanalƒ± (Sadece Saƒüdaki Var)")

# Alt Sƒ±ra: Evri≈üim Sonu√ßlarƒ±
ax[1, 0].imshow(np.abs(r_cikti), cmap='gray')
ax[1, 0].set_title("Kƒ±rmƒ±zƒ± Kanalƒ± Kenarlarƒ±")
ax[1, 1].imshow(np.abs(g_cikti), cmap='gray')
ax[1, 1].set_title("Ye≈üil Kanalƒ± Kenarlarƒ±")

# Final Sonu√ß
ax[1, 2].imshow(nihai_harita, cmap='gray')
ax[1, 2].set_title("Nƒ∞HAƒ∞ HARƒ∞TA (Hepsi Toplandƒ±)")

plt.tight_layout()
plt.show()
```

**1. VGGNet (VGG16 ve VGG19)**

2014 yƒ±lƒ±nda Oxford √úniversitesi Visual Geometry Group tarafƒ±ndan geli≈ütirilmi≈ütir. AlexNet‚Äôin halefi olarak kabul edilir ancak mimari felsefesi √ßok daha d√ºzenlidir.

**Temel Mimari Felsefesi:** VGG'nin en b√ºy√ºk inovasyonu, b√ºy√ºk filtreler yerine (√∂rn: $5 \times 5$ veya $7 \times 7$) ardƒ±≈üƒ±k k√º√ß√ºk ($3 \times 3$) filtreler kullanmasƒ±dƒ±r.

Neden $3 \times 3$?

- ƒ∞ki adet $3 \times 3$ konvol√ºsyon katmanƒ±, bir adet $5 \times 5$ katmanƒ±yla aynƒ± "Receptive Field"a (Algƒ±lama Alanƒ±) sahiptir.

- Ancak, iki katman kullanmak araya iki adet ReLU (doƒürusal olmayan aktivasyon) eklenmesini saƒülar. Bu da modelin daha karma≈üƒ±k √∂zellikleri √∂ƒürenmesine olanak tanƒ±r.

- Parametre sayƒ±sƒ± azalƒ±r: $2 \cdot (3^2) = 18$ parametreye kar≈üƒ±, $1 \cdot (5^2) = 25$ parametre.

**Mimari Detaylarƒ± (VGG16):**

**- Giri≈ü:** $224 \times 224$ RGB g√∂r√ºnt√º.

**- Bloklar:** 5 ana konvol√ºsyon bloƒüu vardƒ±r. Her blokta $3 \times 3$ filtreler kullanƒ±lƒ±r (stride=1, padding=1). Blok sonlarƒ±nda boyut k√º√ß√ºltmek i√ßin Max Pooling ($2 \times 2$, stride=2) uygulanƒ±r.

**- Filtre Sayƒ±larƒ±:** Derinle≈ütik√ße filtre sayƒ±sƒ± artar: $64 \rightarrow 128 \rightarrow 256 \rightarrow 512 \rightarrow 512$.

**- Fully Connected (FC) Katmanlarƒ±:** Sonunda 3 adet FC katmanƒ± bulunur (4096, 4096, 1000 n√∂ron).Dezavantaj: VGG'nin parametrelerinin %80'inden fazlasƒ± bu FC katmanlarƒ±ndadƒ±r (Toplam ~138 Milyon parametre). Bu nedenle model olduk√ßa aƒüƒ±rdƒ±r.

**2. ResNet (Residual Networks)**

2015 yƒ±lƒ±nda Microsoft Research tarafƒ±ndan geli≈ütirildi ve derin aƒülarƒ±n √∂n√ºndeki en b√ºy√ºk engel olan Vanishing Gradient (Kaybolan T√ºrev) problemini √ß√∂zd√º.

**Temel Mimari Felsefesi:** Aƒü derinle≈ütik√ße (√∂rn: 20 katmandan sonra), eƒüitim hatasƒ± azalmak yerine artmaya ba≈ülar. ResNet, Skip Connections (Atlama Baƒülantƒ±larƒ±) veya diƒüer adƒ±yla "Shortcut"lar kullanarak bu sorunu a≈üar.

**Residual Blok Mantƒ±ƒüƒ±:**
Normal bir aƒü $H(x)$ fonksiyonunu √∂ƒürenmeye √ßalƒ±≈üƒ±rken, ResNet $F(x) = H(x) - x$ yani "artƒ±k" (residual) deƒüeri √∂ƒürenmeye √ßalƒ±≈üƒ±r.

$$H(x) = F(x) + x$$

Burada $x$ giri≈ü verisidir. $x$, konvol√ºsyon katmanlarƒ±nƒ± atlayarak (identity mapping) √ßƒ±kƒ±≈üa doƒürudan eklenir.

**Neden √áalƒ±≈üƒ±r?** 
Geri yayƒ±lƒ±m (Backpropagation) sƒ±rasƒ±nda t√ºrev hesaplanƒ±rken, toplama i≈ülemi sayesinde gradyanlar $x$ yolu √ºzerinden bozulmadan geriye akabilir. Bu, 152 katmanlƒ± (ResNet-152) aƒülarƒ±n bile eƒüitilmesini m√ºmk√ºn kƒ±lar.

**Bottleneck Mimarisi (ResNet-50 ve √ºzeri):**

Hesaplama maliyetini d√º≈ü√ºrmek i√ßin ResNet-50 ve √ºzeri modellerde "Bottleneck" bloƒüu kullanƒ±lƒ±r. Bu blok 3 katmandan olu≈üur:

- $1 \times 1$ Conv: Boyut (kanal sayƒ±sƒ±) azaltma (√ñrn: 256 -> 64).

- $3 \times 3$ Conv: √ñzellik √ßƒ±karma.

- $1 \times 1$ Conv: Boyutu tekrar y√ºkseltme (√ñrn: 64 -> 256).

Bu yapƒ±, $3 \times 3$ konvol√ºsyonun daha d√º≈ü√ºk boyutlu uzayda √ßalƒ±≈ümasƒ±nƒ± saƒülayarak i≈ülem g√ºc√ºnden tasarruf eder.

**3. Inception (GoogLeNet)**

Google tarafƒ±ndan geli≈ütirilen bu mimari, aƒüƒ±n geni≈ülemesine (width) odaklanƒ±r.

**Temel Mimari Felsefesi:**

Bir katmanda hangi filtre boyutunu kullanmalƒ±yƒ±z? $1 \times 1$ mi, $3 \times 3$ m√º, yoksa $5 \times 5$ mi? Inception mod√ºl√º "Hepsini kullan" der.

- $3 \times 3$ Max Pooling uygulanƒ±r ve sonu√ßlar kanal boyutunda (depth) birle≈ütirilir (concatenation).

**4. MobileNet (Verimlilik Odaklƒ±)**

Mobil ve g√∂m√ºl√º cihazlar i√ßin tasarlanmƒ±≈ütƒ±r. Ana hedefi doƒüruluƒüu korurken i≈ülem y√ºk√ºn√º (FLOPs) ve model boyutunu k√º√ß√ºltmektir.

Temel ƒ∞novasyon: Depthwise Separable Convolution Standart konvol√ºsyonu iki a≈üamaya b√∂ler:

- **Depthwise Convolution:** Her giri≈ü kanalƒ± i√ßin ayrƒ± bir filtre uygular (Mekansal √∂zellikleri √∂ƒürenir).

- **Pointwise Convolution ($1 \times 1$):** Kanal √ßƒ±ktƒ±larƒ±nƒ±n lineer kombinasyonunu olu≈üturur (Kanallar arasƒ± ili≈ükiyi √∂ƒürenir).

$$\text{Maliyet Azalmasƒ±} \approx \frac{1}{N} + \frac{1}{D_k^2}$$

Burada $N$ filtre sayƒ±sƒ±, $D_k$ ise filtre boyutudur (genelde 3). Bu teknik, standart konvol√ºsyona g√∂re 8-9 kat daha az i≈ülem g√ºc√º gerektirir

**5. EfficientNet (SOTA - State of the Art)**

Google Brain tarafƒ±ndan geli≈ütirilen bu model, √∂l√ßekleme (scaling) problemine sistematik bir √ß√∂z√ºm getirir.

Compound Scaling (Bile≈üik √ñl√ßekleme): √ñnceki modeller performansƒ± artƒ±rmak i√ßin sadece bir boyutu artƒ±rƒ±yordu:

- ResNet $\rightarrow$ Daha derin (Depth)

- WideResNet $\rightarrow$ Daha geni≈ü (Width/Channel)

- Y√ºksek √á√∂z√ºn√ºrl√ºk $\rightarrow$ Daha b√ºy√ºk giri≈ü g√∂rseli (Resolution)

EfficientNet, bu √º√ß parametreyi ($depth, width, resolution$) belirli bir katsayƒ± ($\phi$) ile dengeli bir ≈üekilde aynƒ± anda artƒ±rƒ±r. Bu optimize edilmi≈ü katsayƒ±lar Neural Architecture Search (NAS) ile bulunmu≈ütur.

**√ñrnek** *A≈üaƒüƒ±da pre-trained model kullanƒ±larak bir arabanƒ±n √ße≈üitli kanallardaki g√∂r√ºnt√º i≈ülemesi verilmi≈ütir (L√ºtfen kodun sonundaki √ßƒ±karƒ±mlarƒ± okuyunuz)*

```python
# visualize_feature_maps.py
# Gereksinimler: torch, torchvision, matplotlib, pillow, numpy
# √áalƒ±≈ütƒ±rma: python visualize_feature_maps.py [opsiyonel_resim_yolu]

import sys
import os
from PIL import Image, ImageDraw
import numpy as np
import torch
import torchvision.transforms as T
import torchvision.models as models
import matplotlib.pyplot as plt


def make_synthetic_car_image(size=(224,224)):
    img = Image.new('RGB', size, (200, 220, 255)) # (224,224) piksel boyutunda
    draw = ImageDraw.Draw(img)
    w,h = size
    # araba g√∂vdesi
    draw.rectangle([w*0.2, h*0.5, w*0.8, h*0.7], fill=(180,30,30))
    # cam
    draw.rectangle([w*0.55, h*0.52, w*0.72, h*0.62], fill=(120,200,240))
    # tekerlekler
    draw.ellipse([w*0.25-20, h*0.7, w*0.25+20, h*0.86], fill=(30,30,30))
    draw.ellipse([w*0.75-20, h*0.7, w*0.75+20, h*0.86], fill=(30,30,30))
    return img

def load_image(path=None, size=(224,224)):
    if path and os.path.exists(path):
        img = Image.open(path).convert('RGB')
        img = img.resize(size)
    else:
        img = make_synthetic_car_image(size)
    return img

def preprocess(img):
    transform = T.Compose([
        T.ToTensor(),
        T.Normalize(mean=[0.485,0.456,0.406], std=[0.229,0.224,0.225])
    ])
    return transform(img).unsqueeze(0)  # batch dim

def get_model(pretrained=True):
    try:
        model = models.vgg16(pretrained=pretrained).features  # only conv part
        model.eval()
        print(f"Model: VGG16.features loaded (pretrained={pretrained})")
        return model
    except Exception as e:
        print("Pretrained model y√ºklenemedi, rastgele aƒüƒ± kullanƒ±yorum. Hata:", e)
        # Basit birka√ß konvol√ºsyonlu k√º√ß√ºk aƒü (√∂rnek ama√ßlƒ±)
        small = torch.nn.Sequential(
            torch.nn.Conv2d(3, 8, 3, padding=1), torch.nn.ReLU(),
            torch.nn.Conv2d(8, 16, 3, padding=1), torch.nn.ReLU(),
            torch.nn.Conv2d(16, 32, 3, padding=1), torch.nn.ReLU()
        )
        small.eval()
        return small
    
'''
üöÄ Neden pretrained model kullanƒ±yoruz?

√á√ºnk√º sƒ±fƒ±rdan model eƒüitmek:

- √ßok fazla veri ister (binlerce‚Äìmilyonlarca fotoƒüraf)

- √ßok fazla hesap g√ºc√º ister

- √ßok zaman ister

Bu y√ºzden pretrained model alƒ±p:

- ya doƒürudan tahmin yaparƒ±z (inference)

- ya da √ºst katmanlarƒ± deƒüi≈ütirip kendi veri setimize g√∂re yeniden eƒüitiriz (fine-tuning)
'''

def extract_feature_maps(model, x, layer_idxs=None):
    """
    model: torch.nn.Sequential ya da benzeri, katman katman iterasyona izin veren
    x: input tensor
    layer_idxs: hangi katman indekslerinde feature map almak istediƒüin listesi (None ise birka√ß ilki alƒ±r)
    """
    activations = {}
    out = x
    for i, layer in enumerate(model):
        out = layer(out)
        if layer_idxs is None:
            # default: kaydet ilk, orta ve son conv √ßƒ±ktƒ±larƒ± (varsa)
            pass
        # kaydet konv √ßƒ±ktƒ±larƒ± (2D conv -> tensor shape BxCxHxW)
        if isinstance(layer, torch.nn.modules.conv.Conv2d):
            activations[i] = out.detach().cpu()
    if layer_idxs:
        # filtreleyelim
        activations = {k:v for k,v in activations.items() if k in layer_idxs}
    return activations

def plot_feature_maps(activations, out_dir="feature_maps", max_channels=16):
    """
    activations: dict[layer_idx] = tensor BxCxHxW
    √áƒ±ktƒ±larƒ± /feature_maps klas√∂r√ºne kaydeder ve ekranda g√∂sterir.
    """
    os.makedirs(out_dir, exist_ok=True)
    for layer_idx, tensor in activations.items():
        # tensor shape: B x C x H x W
        b,c,h,w = tensor.shape
        # alttaki √∂rnek sadece ilk batch i√ßin
        fmap = tensor[0]
        channels = min(c, max_channels)
        ncols = int(np.ceil(np.sqrt(channels)))
        nrows = int(np.ceil(channels / ncols))
        fig, axes = plt.subplots(nrows, ncols, figsize=(ncols*1.8, nrows*1.8))
        axes = np.array(axes).reshape(-1)
        vmin = float(fmap.min().item())
        vmax = float(fmap.max().item())
        for i in range(len(axes)):
            ax = axes[i]
            ax.axis('off')
            if i < channels:
                img = fmap[i].numpy()
                # normalize per-channel for visualization
                im = (img - img.min()) / (img.max() - img.min() + 1e-9)
                ax.imshow(im, cmap='gray')
            else:
                ax.imshow(np.zeros((10,10)), cmap='gray')
        plt.suptitle(f"Layer {layer_idx} ‚Äî channels shown: {channels} (shape {c}x{h}x{w})")
        plt.tight_layout()
        savepath = os.path.join(out_dir, f"layer_{layer_idx}.png")
        plt.savefig(savepath, dpi=150)
        print("Saved", savepath)
        plt.show()

def main():
    img_path = sys.argv[1] if len(sys.argv) > 1 else None
    pil = load_image(img_path, size=(224,224))
    display_choice = True
    # G√∂r√ºnt√ºy√º g√∂ster
    plt.figure(figsize=(4,4))
    plt.imshow(pil)
    plt.axis('off')
    plt.title("Input image")
    plt.show()

    x = preprocess(pil)
    model = get_model(pretrained=True)
    activations = extract_feature_maps(model, x, layer_idxs=None)
    # Hangi layer'lerin √ßƒ±ktƒ±sƒ±nƒ± tutuyoruz bakalƒ±m
    print("Extracted activation layers:", list(activations.keys()))
    # G√∂rselle≈ütir (katman ba≈üƒ±na en fazla 16 kanal g√∂ster)
    plot_feature_maps(activations, out_dir="feature_maps", max_channels=16)
    print("Tamamlandƒ±. feature_maps klas√∂r√ºn√º kontrol et.")

if __name__ == "__main__":
    main()
```

- ƒ∞lk Katmanlar (Layer 0-5 civarƒ±): Bu √ßƒ±ktƒ±lar genellikle kenarlarƒ± (edges), renk ge√ßi≈ülerini ve basit dokularƒ± g√∂sterir. Aracƒ±n dƒ±≈ü hatlarƒ±nƒ± √ßok net g√∂rebilirsin.

- Orta Katmanlar: Model artƒ±k "yuvarlak ≈üekil", "ƒ±zgara deseni" gibi par√ßalarƒ± birle≈ütirmeye ba≈ülar. Kodundaki araba √∂rneƒüinde tekerleklerin veya pencerelerin belirli filtrelerde parladƒ±ƒüƒ±nƒ± (beyazla≈ütƒ±ƒüƒ±nƒ±) g√∂rebilirsin.

- Derin Katmanlar (Sonlara doƒüru): G√∂r√ºnt√º artƒ±k insan g√∂z√º i√ßin anlamsƒ±zla≈üƒ±r, pikselle≈üir ve soyutla≈üƒ±r. Ancak burasƒ± modelin "Bu bir arabadƒ±r" kararƒ± verdiƒüi yerdir. Burada model artƒ±k piksel deƒüil, "kavram" (concept) g√∂r√ºr.

- √áƒ±ktƒ± resimlerine dikkatlice bak. Eƒüer bazƒ± kareler (kanallar) tamamen siyahsa, bu o filtrenin hi√ß aktive olmadƒ±ƒüƒ±nƒ± (output = 0) g√∂sterir.

- Ne anlama gelir? O filtre √∂ƒürenme s√ºrecinde √∂lm√º≈ü olabilir (Dead ReLU sorunu) veya o filtre bu spesifik resim i√ßin gereksizdir.

*Neden en derin katman en garip verileri i√ßerirken bu bir arabadƒ±r anlamƒ±na gelir*

- ƒ∞lk katmanlar "G√∂r√ºnt√º neye benziyor?" sorusuna cevap verirken, son katmanlar "G√∂r√ºnt√ºde ne var?" sorusuna cevap verir.

- **Giri≈ü (Pikseller):** HD bir araba fotoƒürafƒ±nda milyonlarca detay vardƒ±r (√ßizikler, ƒ±≈üƒ±k yansƒ±masƒ±, asfalttaki ta≈ülar). Bu, bilgisayar i√ßin √ßok fazla "g√ºr√ºlt√º" demektir.

- **Derin Katman (Kavramlar):** Model katmanlarƒ± ilerledik√ße, gereksiz detaylarƒ± (arabanƒ±n rengi kƒ±rmƒ±zƒ± mƒ± bordo mu, g√∂lgesi nereye d√º≈üm√º≈ü) atar. Geriye sadece saf √∂z√º bƒ±rakƒ±r.

Benzetme: Bir kitabƒ± okuduƒüunu d√º≈ü√ºn.

- **ƒ∞lk katman (G√∂zlerin):** Harflerin yazƒ± tipini, kaƒüƒ±dƒ±n rengini, harflerin kƒ±vrƒ±mlarƒ±nƒ± g√∂r√ºr. (Detaylƒ± ama anlamsƒ±z g√∂rsel veri).

- **Derin katman (Beynin):** Kitabƒ± bitirdiƒüinde yazƒ± tipini hatƒ±rlamazsƒ±n, sadece hikayeyi hatƒ±rlarsƒ±n. "Kahraman √∂ld√º" bilgisi √ßok basittir, tek bir c√ºmledir, ama arkasƒ±nda y√ºzlerce sayfalƒ±k detayƒ±n √∂zeti vardƒ±r. ƒ∞≈üte son katmandaki o "basit" kareler, bu hikayenin √∂zetidir.

**Uzaysal Bilginin Kaybƒ± (Spatial Invariance)**

- Modelin bir arabayƒ± tanƒ±masƒ± i√ßin arabanƒ±n resmin tam ortasƒ±nda, saƒüƒ±nda veya solunda olmasƒ± fark etmemelidir.

- Derin katmanlar bilerek "bulanƒ±kla≈üƒ±r" ve konumsal netliƒüi kaybeder.

- **B√∂ylece model ≈üunu diyebilir:** "Arabanƒ±n tam olarak hangi milimetrede durduƒüu umurumda deƒüil, resmin o b√∂lgesinde g√º√ßl√º bir 'araba olma' sinyali alƒ±yorum."

**√ñzetle**

- O g√∂rd√ºƒü√ºn 7x7 veya 14x14'l√ºk bulanƒ±k karelerin her biri, aslƒ±nda y√ºzlerce sorunun cevabƒ±nƒ± ta≈üƒ±yan s√ºper yoƒüun bir bilgi paketidir.

- **G√∂r√ºn√º≈ü:** Bulanƒ±k, d√º≈ü√ºk √ß√∂z√ºn√ºrl√ºkl√º, basit.

- **ƒ∞√ßerik:** "Bu b√∂lgede metalik bir doku, tekerlek benzeri yuvarlaklƒ±k ve cam y√ºzey var -> Olasƒ±lƒ±k %90 Araba."

**Generative Models**

VAE (Variational Autoencoder - Varyasyonel Otokodlayƒ±cƒ±) ve GAN (Generative Adversarial Network - √áeki≈ümeli √úretici Aƒü), derin √∂ƒürenme d√ºnyasƒ±nda "Generative AI" (√úretken Yapay Zeka) olarak bilinen alanƒ±n en temel iki mimarisidir. Her ikisi de var olan veriyi (√∂rneƒüin insan y√ºzleri) analiz ederek, eƒüitim setinde olmayan ama ona √ßok benzeyen yeni veriler √ºretmek i√ßin kullanƒ±lƒ±r.

Aralarƒ±ndaki temel fark, √∂ƒürenme y√∂ntemlerindedir: VAE olasƒ±lƒ±ksal (probabilistic) bir yakla≈üƒ±m kullanƒ±rken, GAN oyun teorisine dayalƒ± rekabet√ßi bir yakla≈üƒ±m kullanƒ±r.

ƒ∞≈üte bu iki teknolojinin teknik detaylarƒ± ve kar≈üƒ±la≈ütƒ±rmasƒ±:

**1. VAE (Variational Autoencoder)**

VAE, klasik "Autoencoder" yapƒ±sƒ±nƒ±n √ºzerine in≈üa edilmi≈ü istatistiksel bir modeldir. Klasik bir Autoencoder veriyi sƒ±kƒ±≈ütƒ±rƒ±p (encode) sonra tekrar a√ßarak (decode) aynƒ±sƒ±nƒ± olu≈üturmaya √ßalƒ±≈üƒ±rken, VAE verinin daƒüƒ±lƒ±mƒ±nƒ± (distribution) √∂ƒürenmeye √ßalƒ±≈üƒ±r.

- **Nasƒ±l √áalƒ±≈üƒ±r?**
1. **Encoder (Kodlayƒ±cƒ±):** Girdi verisini (√∂rneƒüin bir resim) alƒ±r ve onu bir "Gizli Uzay"a (Latent Space) sƒ±kƒ±≈ütƒ±rƒ±r. Ancak VAE, veriyi tek bir noktaya sƒ±kƒ±≈ütƒ±rmak yerine, bir olasƒ±lƒ±k daƒüƒ±lƒ±mƒ±na (genellikle Gauss daƒüƒ±lƒ±mƒ±: Ortalama $\mu$ ve Varyans $\sigma$) d√∂n√º≈üt√ºr√ºr.

2. **Latent Space (Gizli Uzay):** Buradan rastgele bir √∂rneklem (sample) alƒ±nƒ±r. Bu, modelin sadece ezberlemesini deƒüil, verinin yapƒ±sƒ±nƒ± √∂ƒürenmesini saƒülar.

3. **Decoder (Kod √á√∂z√ºc√º):** √ñrneklenen bu noktayƒ± alƒ±r ve onu tekrar orijinal veriye benzer bir g√∂r√ºnt√ºye d√∂n√º≈üt√ºr√ºr.

- **Matematiksel Arka Plan:** VAE, ELBO (Evidence Lower Bound) adƒ± verilen bir deƒüeri optimize eder. Kayƒ±p fonksiyonunda iki terim vardƒ±r:

1. Reconstruction Loss: Girdi ve √ßƒ±ktƒ± ne kadar benziyor?

2. KL Divergence: √ñƒürenilen daƒüƒ±lƒ±m, normal daƒüƒ±lƒ±ma ne kadar benziyor?

- **Sonu√ß:** VAE'ler genellikle daha bulanƒ±k (blurry) ama s√ºrekli ge√ßi≈ülere sahip g√∂r√ºnt√ºler √ºretir. ƒ∞ki y√ºz arasƒ±nda "morphing" (d√∂n√º≈ü√ºm) yapmak istediƒüinizde VAE √ßok ba≈üarƒ±lƒ±dƒ±r √ß√ºnk√º gizli uzayƒ± s√ºreklidir.

**2. GAN (Generative Adversarial Network)**

GAN, birbiriyle sava≈üan iki sinir aƒüƒ±ndan olu≈üur. Bu yapƒ±, bir "kalpazan ve polis" oyununa benzetilebilir.

- Nasƒ±l √áalƒ±≈üƒ±r?

1. **Generator (√úretici - Kalpazan):** Rastgele g√ºr√ºlt√ºden (random noise) ba≈ülar ve ger√ßek√ßi bir veri (resim) √ºretmeye √ßalƒ±≈üƒ±r. Amacƒ± Discriminator'ƒ± kandƒ±rmaktƒ±r.

2. **Discriminator (Ayƒ±rt Edici - Polis):** Kendisine gelen verinin "Ger√ßek" (veri setinden) mi yoksa "Sahte" (Generator'dan) mi olduƒüunu anlamaya √ßalƒ±≈üƒ±r.

![scatter plot](https://i.imgur.com/MAkZEjV.jpeg)
 - **S√ºre√ß:**
 
- Generator sahte resim yapar -> Discriminator "Bu sahte" der.

- Generator hatasƒ±ndan ders alƒ±r, daha iyisini yapar.

- Discriminator daha dikkatli bakmayƒ± √∂ƒürenir.

- Bu d√∂ng√º, Discriminator artƒ±k sahte ve ger√ßek arasƒ±ndaki farkƒ± anlayamayacak duruma gelene kadar (Nash Dengesi) devam eder.

Sonu√ß: GAN'lar √ßok daha keskin, ger√ßek√ßi ve detaylƒ± g√∂r√ºnt√ºler √ºretir. Ancak eƒüitimi zordur; dengesizlik olu≈üursa model hi√ßbir ≈üey √∂ƒürenemeyebilir (Mode Collapse).

| √ñzellik             | VAE (Varyasyonel Otokodlayƒ±cƒ±)                          | GAN (√áeki≈ümeli √úretici Aƒü)                              |
|--------------------|--------------------------------------------------------|--------------------------------------------------------|
| Temel Mantƒ±k        | Olasƒ±lƒ±k daƒüƒ±lƒ±mƒ±nƒ± √∂ƒürenir ve √∂rneklem alƒ±r.         | ƒ∞ki aƒüƒ±n rekabeti (Minimax oyunu).                    |
| G√∂r√ºnt√º Kalitesi    | Genellikle daha bulanƒ±k ve d√º≈ü√ºk detaylƒ±dƒ±r.          | √áok keskin, ger√ßek√ßi ve y√ºksek detaylƒ±dƒ±r.           |
| Eƒüitim Kararlƒ±lƒ±ƒüƒ±  | Eƒüitimi daha kolay ve kararlƒ±dƒ±r (Converge eder).      | Eƒüitimi zordur, kararsƒ±zla≈üabilir (Unstable).        |
| Latent Space (Gizli Uzay) | S√ºreklidir; veriler arasƒ± yumu≈üak ge√ßi≈ü yapƒ±labilir. | Genellikle karma≈üƒ±ktƒ±r; yumu≈üak ge√ßi≈ü garantisi yoktur. |
| Kullanƒ±m Alanƒ±      | Veri sƒ±kƒ±≈ütƒ±rma, g√ºr√ºlt√º giderme, anomali tespiti.   | Deepfake, g√∂r√ºnt√ºden g√∂r√ºnt√ºye √ßeviri, s√ºper √ß√∂z√ºn√ºrl√ºk. |


Hangisini Kullanmalƒ±sƒ±nƒ±z?
Eƒüer amacƒ±nƒ±z verinin yapƒ±sƒ±nƒ± anlamak, veriyi analiz etmek veya iki veri arasƒ±nda yumu≈üak ge√ßi≈üler yapmaksa (√∂rneƒüin bir ayakkabƒ± tasarƒ±mƒ±nƒ± yava≈ü√ßa bir bota d√∂n√º≈üt√ºrmek), VAE daha uygundur.

Eƒüer amacƒ±nƒ±z insan g√∂z√ºn√ºn ayƒ±rt edemeyeceƒüi kadar ger√ßek√ßi ve keskin g√∂r√ºnt√ºler olu≈üturmaksa (√∂rneƒüin olmayan insan y√ºzleri yaratmak), GAN daha iyi bir tercihtir.

**√ñrnek** *MNIST veriseti √ºzerinden VAE ve GAN √∂rneƒüi*

- **Encoder:** G√∂r√ºnt√ºy√º alƒ±r, bir ortalama ($\mu$) ve varyans (logvar) vekt√∂r√ºne sƒ±kƒ±≈ütƒ±rƒ±r.

- **Reparameterize:** Bu $\mu$ ve logvar deƒüerlerini kullanarak "√∂rneklem" (sample) alƒ±r.

- **Decoder:** √ñrneklenen veriyi tekrar g√∂r√ºnt√ºye √ßevirir.

*Yani g√∂r√ºnt√ºden tekrardan veri √ºretebilmek i√ßin istatistiksel ara√ßlar ile sistematik veri elde ederiz, herhangi bir piksel verisi se√ßseydik rastgele √∂rnekler olu≈üturmu≈ü olurduk ancak encoder mantƒ±ƒüƒ± ile s√ºrekli olmasƒ±nƒ± saƒülamak yani a≈üaƒüƒ±da kod √ßƒ±ktƒ±larƒ±nƒ± incelediƒüinizde g√∂receƒüiniz CNN ile √ºretilmi≈ü sonu√ßlara yakƒ±n √∂rneklerin (bulanƒ±k sayƒ±, 3 ve 8 karƒ±≈üƒ±mƒ± bir ≈üey, anlamlƒ± sayƒ± ifade etmiyor ama eldeki verilerle √ºretildiƒüi belli) olduƒüunu g√∂rebilirsiniz*

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
import torch.nn.functional as F

# 1. Ayarlar ve Hiperparametreler
BATCH_SIZE = 100
EPOCHS = 10
LR = 1e-3
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Veri Setini Hazƒ±rlama (MNIST)
# VAE giri≈ü olarak d√ºzle≈ütirilmi≈ü vekt√∂r (28x28=784) bekler.
transform = transforms.Compose([
    transforms.ToTensor(),
])

train_dataset = datasets.MNIST(root='./data', train=True, transform=transform, download=True)
train_loader = DataLoader(dataset=train_dataset, batch_size=BATCH_SIZE, shuffle=True)

# 2. VAE Model Mimarisi
class VAE(nn.Module):
    def __init__(self):
        super(VAE, self).__init__()
        
        # --- ENCODER ---
        # Girdi: 784 (28x28 piksel) -> Gizli Katman: 400
        self.fc1 = nn.Linear(784, 400)
        
        # Gizli uzay i√ßin (Latent Space) ortalama (mu) ve log-varyans (logvar) katmanlarƒ±
        # Latent boyutu (z_dim) = 20 olsun
        self.fc21 = nn.Linear(400, 20) # Mean (Ortalama)
        self.fc22 = nn.Linear(400, 20) # Log-Variance (Varyansƒ±n logaritmasƒ±)

        # --- DECODER ---
        # Latent (20) -> Gizli Katman: 400
        self.fc3 = nn.Linear(20, 400)
        # Gizli Katman -> √áƒ±ktƒ±: 784
        self.fc4 = nn.Linear(400, 784)

    def encode(self, x):
        """Girdiyi alƒ±r ve latent daƒüƒ±lƒ±m parametrelerini (mu, logvar) d√∂nd√ºr√ºr."""
        h1 = F.relu(self.fc1(x))
        return self.fc21(h1), self.fc22(h1)

    def reparameterize(self, mu, logvar):
        """
        Reparameterization Trick:
        Geriye yayƒ±lƒ±m (backpropagation) yapabilmek i√ßin rastgeleliƒüi
        modelin dƒ±≈üƒ±na ta≈üƒ±rƒ±z.
        z = mu + std * epsilon
        """
        std = torch.exp(0.5 * logvar) # Log varyansƒ± standart sapmaya √ßevir
        eps = torch.randn_like(std)   # Standart normal daƒüƒ±lƒ±mdan rastgele g√ºr√ºlt√º (epsilon)
        return mu + eps * std
    
    '''
    Sistemin kaotik deƒüil de "sistematik" olmasƒ±nƒ± saƒülayan form√ºl mu + eps * std

    Bu kƒ±smƒ± anlamak olduk√ßa zorlayabilir dolayƒ±sƒ±yla √ßƒ±ktƒ±nƒ±n girdi resimlere benzer ama bulanƒ±k olmasƒ±nƒ±n sebebinin burasƒ± olduƒüunu s√∂yleyebiliriz
    '''

    def decode(self, z):
        """Latent vekt√∂r√º (z) tekrar g√∂r√ºnt√ºye √ßevirir."""
        h3 = F.relu(self.fc3(z))
        return torch.sigmoid(self.fc4(h3)) # √áƒ±ktƒ±yƒ± 0-1 arasƒ±na sƒ±kƒ±≈ütƒ±r (Piksel deƒüerleri i√ßin)

    def forward(self, x):
        mu, logvar = self.encode(x.view(-1, 784)) # G√∂r√ºnt√ºy√º d√ºzle≈ütir
        z = self.reparameterize(mu, logvar)
        return self.decode(z), mu, logvar

# Modeli ba≈ülat
model = VAE().to(DEVICE)
optimizer = optim.Adam(model.parameters(), lr=LR)

# 3. Kayƒ±p Fonksiyonu (Loss Function)
# VAE Loss = Reconstruction Loss + KL Divergence
def loss_function(recon_x, x, mu, logvar):
    # 1. Reconstruction Loss: √áƒ±ktƒ± ile Orijinal arasƒ±ndaki fark (Binary Cross Entropy)
    BCE = F.binary_cross_entropy(recon_x, x.view(-1, 784), reduction='sum')

    # 2. KL Divergence: √ñƒürenilen daƒüƒ±lƒ±mƒ±n Normal Daƒüƒ±lƒ±mdan ne kadar saptƒ±ƒüƒ±
    # Form√ºl: -0.5 * sum(1 + log(sigma^2) - mu^2 - sigma^2)
    KLD = -0.5 * torch.sum(1 + logvar - mu.pow(2) - logvar.exp())

    return BCE + KLD

# 4. Eƒüitim D√∂ng√ºs√º
print(f"Eƒüitim {DEVICE} √ºzerinde ba≈ülƒ±yor...")
model.train()

for epoch in range(EPOCHS):
    train_loss = 0
    for batch_idx, (data, _) in enumerate(train_loader):
        data = data.to(DEVICE)
        
        optimizer.zero_grad()
        
        # Forward pass
        recon_batch, mu, logvar = model(data)
        
        # Loss hesapla
        loss = loss_function(recon_batch, data, mu, logvar)
        
        # Backward pass ve optimizasyon
        loss.backward()
        train_loss += loss.item()
        optimizer.step()
        
    print(f'Epoch: {epoch+1} | Ortalama Loss: {train_loss / len(train_loader.dataset):.4f}')

print("Eƒüitim tamamlandƒ±!")
```

- **Reconstruction (Yeniden Olu≈üturma):** Ger√ßek sayƒ±lar ile modelin onlarƒ± nasƒ±l g√∂rd√ºƒü√ºn√º kƒ±yaslayacak.

- **Generation (√úretim):** Rastgele g√ºr√ºlt√ºden (noise) hi√ß var olmayan yeni rakamlar √ºretecek.

```python
import matplotlib.pyplot as plt
import torchvision.utils as vutils

# Modeli deƒüerlendirme moduna al (Dropout vb. kapatƒ±r)
model.eval()

def visualize_results():
    with torch.no_grad():
        # 1. RECONSTRUCTION (Orijinal vs VAE √áƒ±ktƒ±sƒ±)
        # Test i√ßin veri setinden bir batch alalƒ±m
        data_iter = iter(train_loader)
        images, _ = next(data_iter)
        images = images.to(DEVICE)

        # Modeli √ßalƒ±≈ütƒ±r
        recon_images, _, _ = model(images)

        # ƒ∞lk 8 resmi yan yana koyalƒ±m
        # Orijinaller (√úst Sƒ±ra)
        original_grid = vutils.make_grid(images[:8], nrow=8, padding=2, normalize=True)
        # VAE √áƒ±ktƒ±larƒ± (Alt Sƒ±ra)
        recon_grid = vutils.make_grid(recon_images.view(-1, 1, 28, 28)[:8], nrow=8, padding=2, normalize=True)

        plt.figure(figsize=(15, 5))
        plt.subplot(1, 2, 1)
        plt.axis("off")
        plt.title("Orijinal Resimler")
        plt.imshow(original_grid.cpu().numpy().transpose((1, 2, 0)))

        plt.subplot(1, 2, 2)
        plt.axis("off")
        plt.title("VAE Tarafƒ±ndan Yeniden Olu≈üturulan")
        plt.imshow(recon_grid.cpu().numpy().transpose((1, 2, 0)))
        plt.show()

        # 2. GENERATION (Yoktan Rakam √úretme)
        # Latent Space'den (z) rastgele √∂rneklem alalƒ±m (Normal Daƒüƒ±lƒ±m)
        # 64 tane rastgele vekt√∂r √ºretelim
        z = torch.randn(64, 20).to(DEVICE) # 20 bizim latent boyumuzdu

        # Decoder'a sadece bu g√ºr√ºlt√ºy√º verelim
        generated_images = model.decode(z)
        generated_images = generated_images.view(-1, 1, 28, 28)

        # √úretilenleri √ßizelim
        gen_grid = vutils.make_grid(generated_images, nrow=8, padding=2, normalize=True)

        plt.figure(figsize=(8, 8))
        plt.axis("off")
        plt.title("VAE ile Rastgele G√ºr√ºlt√ºden √úretilen Rakamlar")
        plt.imshow(gen_grid.cpu().numpy().transpose((1, 2, 0)))
        plt.show()

visualize_results()
```

**√áƒ±ktƒ±da Ne G√∂receksiniz?**

**Reconstruction Kƒ±smƒ±nda:**

- Solda net orijinal rakamlar, saƒüda ise biraz daha bulanƒ±k (blurry) ama okunabilir rakamlar g√∂receksiniz.

- Neden bulanƒ±k? VAE, piksellerin kesin deƒüerini deƒüil, olasƒ±lƒ±k daƒüƒ±lƒ±mƒ±nƒ± √∂ƒürendiƒüi i√ßin (ve MSE/BCE loss kullandƒ±ƒüƒ± i√ßin) √ßƒ±ktƒ±larƒ± ortalama bir deƒüerde tutar, bu da g√∂r√ºnt√ºy√º yumu≈üatƒ±r.

**Generation Kƒ±smƒ±nda:**

- Tamamen makine tarafƒ±ndan uydurulmu≈ü rakamlar g√∂receksiniz. Bunlar veri setindeki rakamlar deƒüil, modelin "rakam nedir" sorusuna verdiƒüi cevaptƒ±r.

- Bazƒ± ≈üekiller tam bir rakama benzemeyebilir veya iki rakamƒ±n karƒ±≈üƒ±mƒ± gibi g√∂r√ºnebilir (√∂rneƒüin 3 ile 8 arasƒ± bir ≈üekil). Bu, VAE'nin s√ºrekli gizli uzay (continuous latent space) √∂zelliƒüinin kanƒ±tƒ±dƒ±r.

*A≈üaƒüƒ±daki kodu √ßalƒ±≈ütƒ±rƒ±n. Dikkat: GAN eƒüitimi daha uzundur ve daha hassastƒ±r. Sonu√ßlarƒ± g√∂rmek i√ßin en az 20-30 epoch sabretmek gerekebilir, ancak 50 epoch civarƒ±nda √ßok net rakamlar olu≈üur. (Kod √ßƒ±ktƒ±larƒ± hazƒ±rda var eƒüer g√∂z√ºkmezse tekrardan √ßalƒ±≈ütƒ±rƒ±n)*

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
import torchvision.utils as vutils
import matplotlib.pyplot as plt

# 1. Ayarlar
BATCH_SIZE = 100
LR = 0.0002 # GAN'lar y√ºksek √∂ƒürenme hƒ±zƒ±nƒ± sevmez
EPOCHS = 50 # Keskin sonu√ßlar i√ßin biraz s√ºre gerekir
Z_DIM = 100 # G√ºr√ºlt√º vekt√∂r√ºn√ºn boyutu
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Veri Hazƒ±rlƒ±ƒüƒ± (Normalize: -1 ile 1 arasƒ±)
# GAN'larda Tanh aktivasyonu kullanƒ±ldƒ±ƒüƒ± i√ßin veriyi de -1, 1 arasƒ±na √ßekiyoruz.
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])

dataset = datasets.MNIST(root='./data', train=True, transform=transform, download=True)
loader = DataLoader(dataset, batch_size=BATCH_SIZE, shuffle=True)

# 2. Modeller

# --- Generator (√úretici) ---
class Generator(nn.Module):
    def __init__(self):
        super(Generator, self).__init__()
        self.model = nn.Sequential(
            nn.Linear(Z_DIM, 256),
            nn.LeakyReLU(0.2), # GAN'larda ReLU yerine LeakyReLU tercih edilir
            nn.Linear(256, 512),
            nn.LeakyReLU(0.2),
            nn.Linear(512, 1024),
            nn.LeakyReLU(0.2),
            nn.Linear(1024, 784), # 28x28 = 784
            nn.Tanh() # √áƒ±ktƒ±yƒ± -1 ile 1 arasƒ±na sƒ±kƒ±≈ütƒ±rƒ±r
        )

    def forward(self, z):
        return self.model(z).view(-1, 1, 28, 28)

# --- Discriminator (Ayƒ±rt Edici) ---
class Discriminator(nn.Module):
    def __init__(self):
        super(Discriminator, self).__init__()
        self.model = nn.Sequential(
            nn.Flatten(),
            nn.Linear(784, 1024),
            nn.LeakyReLU(0.2),
            nn.Dropout(0.3), # Discriminator'ƒ±n ezberlemesini √∂nlemek i√ßin Dropout
            nn.Linear(1024, 512),
            nn.LeakyReLU(0.2),
            nn.Dropout(0.3),
            nn.Linear(512, 256),
            nn.LeakyReLU(0.2),
            nn.Linear(256, 1),
            nn.Sigmoid() # Ger√ßek olma olasƒ±lƒ±ƒüƒ± (0-1)
        )

    def forward(self, img):
        return self.model(img)

# Modelleri Ba≈ülat
generator = Generator().to(DEVICE)
discriminator = Discriminator().to(DEVICE)

# Optimizerlar (Adam genellikle iyidir, beta1=0.5 GAN i√ßin standarttƒ±r)
opt_g = optim.Adam(generator.parameters(), lr=LR, betas=(0.5, 0.999))
opt_d = optim.Adam(discriminator.parameters(), lr=LR, betas=(0.5, 0.999))

# Kayƒ±p Fonksiyonu (BCELoss)
criterion = nn.BCELoss()

# 3. Eƒüitim D√∂ng√ºs√º
print(f"Eƒüitim {DEVICE} √ºzerinde ba≈ülƒ±yor...")

for epoch in range(EPOCHS):
    for batch_idx, (real_imgs, _) in enumerate(loader):
        real_imgs = real_imgs.to(DEVICE)
        batch_len = real_imgs.size(0)

        # Etiketler: Ger√ßekler i√ßin 1, Sahteler i√ßin 0
        real_labels = torch.ones(batch_len, 1).to(DEVICE)
        fake_labels = torch.zeros(batch_len, 1).to(DEVICE)

        # ---------------------
        # A. DISCRIMINATOR Eƒûƒ∞Tƒ∞Mƒ∞
        # Ama√ß: Ger√ßeƒüi ger√ßek, sahteyi sahte olarak tanƒ±.
        # ---------------------
        opt_d.zero_grad()

        # 1. Ger√ßek resimlerle kayƒ±p
        outputs = discriminator(real_imgs)
        d_loss_real = criterion(outputs, real_labels)
        
        # 2. Sahte resimlerle kayƒ±p
        z = torch.randn(batch_len, Z_DIM).to(DEVICE)
        fake_imgs = generator(z)
        outputs = discriminator(fake_imgs.detach()) # .detach() √∂nemli: Generator g√ºncellenmesin
        d_loss_fake = criterion(outputs, fake_labels)

        # Toplam D kaybƒ± ve backprop
        d_loss = d_loss_real + d_loss_fake
        d_loss.backward()
        opt_d.step()

        # ---------------------
        # B. GENERATOR Eƒûƒ∞Tƒ∞Mƒ∞
        # Ama√ß: Discriminator'ƒ± kandƒ±r (Sahte resme "Ger√ßek" dedirt).
        # ---------------------
        opt_g.zero_grad()
        
        # Generator i√ßin Discriminator'ƒ±n √ßƒ±ktƒ±sƒ±nƒ± al (bu sefer detach yok)
        outputs = discriminator(fake_imgs)
        
        # √ñNEMLƒ∞: Hedefimiz "real_labels" (1). √á√ºnk√º kandƒ±rmak istiyoruz.
        g_loss = criterion(outputs, real_labels)
        
        g_loss.backward()
        opt_g.step()

    # ƒ∞lerlemeyi yazdƒ±r
    if (epoch+1) % 10 == 0:
        print(f"Epoch [{epoch+1}/{EPOCHS}] | D Loss: {d_loss.item():.4f} | G Loss: {g_loss.item():.4f}")

print("Eƒüitim tamamlandƒ±!")
```

*Eƒüitim bittiƒüinde, VAE'den farklƒ± olarak √ßok daha keskin ama belki de bazƒ± garip ≈üekiller i√ßeren sonu√ßlar g√∂receksiniz. A≈üaƒüƒ±daki kod ile Generator'ƒ±n neler √∂ƒürendiƒüine bakalƒ±m.*

**VAE ve GAN Arasƒ±ndaki Farkƒ± Nasƒ±l Hissedeceksiniz?**

**Eƒüitim Loss Deƒüerleri:**

- **VAE'de:** Loss s√ºrekli d√º≈üer (150 -> 100).

- **GAN'da:** Loss deƒüerleri dans eder. Discriminator kaybƒ± azalƒ±rken Generator artabilir, sonra tam tersi olabilir. Bu bir denge oyunudur (Nash Equilibrium). Eƒüer biri (D veya G) kaybƒ± sƒ±fƒ±rlarsa (0.0000), oyun biter ve model √∂ƒürenmeyi durdurur (buna "Collapse" denir).

**G√∂r√ºnt√º Kalitesi:**

- Bu kodu √ßalƒ±≈ütƒ±rdƒ±ƒüƒ±nƒ±zda rakamlarƒ±n kenarlarƒ±nƒ±n VAE'ye g√∂re daha keskin beyaz/siyah olduƒüunu, VAE'deki o "gri sisin" burada olmadƒ±ƒüƒ±nƒ± fark edeceksiniz.

*Kavramlar ve teknik kƒ±sƒ±m olduk√ßa zorlayƒ±cƒ± olduƒüu i√ßin bu alanda deneyim kazandƒ±k√ßa anla≈üƒ±lacaktƒ±r dolayƒ±sƒ±yla verilen g√∂revleri yapmanƒ±z ardƒ±ndan zamanla ara≈ütƒ±rma yapmanƒ±z bu kƒ±smƒ± daha iyi anlamanƒ±za yardƒ±mcƒ± olacaktƒ±r, dok√ºman sonunda ara≈ütƒ±rma yapƒ±lan t√ºm kaynaklar sunulmu≈ütur ve modern sinir aƒülarƒ±nƒ± anlamak i√ßin bu dok√ºmanda bulunan kavramlarƒ± ke≈üfetmek ve modern sinir aƒülarƒ±nda neler olduƒüunu g√∂rmek i√ßin √ßok yardƒ±mcƒ± olacaktƒ±r.*

```python
def visualize_gan_results():
    generator.eval()
    with torch.no_grad():
        # Rastgele g√ºr√ºlt√º olu≈ütur
        z = torch.randn(64, Z_DIM).to(DEVICE)
        generated_imgs = generator(z)
        
        # G√∂r√ºnt√ºleri -1,1 aralƒ±ƒüƒ±ndan 0,1 aralƒ±ƒüƒ±na geri √ßek (G√∂rselle≈ütirme i√ßin)
        generated_imgs = (generated_imgs + 1) / 2

        grid = vutils.make_grid(generated_imgs, nrow=8, padding=2, normalize=True)
        
        plt.figure(figsize=(8, 8))
        plt.axis("off")
        plt.title("GAN Tarafƒ±ndan √úretilen Rakamlar")
        plt.imshow(grid.cpu().numpy().transpose((1, 2, 0)))
        plt.show()

visualize_gan_results()
```

*Videodaki ikinci projeyi a≈üaƒüƒ±daki kod satƒ±rƒ±na yapƒ±nƒ±z* [Dataset](https://www.kaggle.com/datasets/andrewmvd/animal-faces)

```python
# Image Classification - Animal Faces (PyTorch transfer learning, AFHQ layout)
# Expects dataset at './animal-faces/afhq/train' and './animal-faces/afhq/val'
import os
import pathlib

try:
    import torch
    import torch.nn as nn
    import torch.optim as optim
    from torch.utils.data import DataLoader
    from torchvision import datasets, transforms, models
    import matplotlib.pyplot as plt
    import numpy as np
except Exception as e:
    print("Gerekli paketler eksik. L√ºtfen 'pip install torch torchvision matplotlib numpy pillow' komutunu √ßalƒ±≈ütƒ±rƒ±n.")
    raise

ROOT = pathlib.Path("./animal-faces/afhq")
TRAIN_DIR = ROOT / "train"
VAL_DIR = ROOT / "val"

if not TRAIN_DIR.exists() or not VAL_DIR.exists():
    print(f"Beklenen veri klas√∂rleri bulunamadƒ±:\n  {TRAIN_DIR}\n  {VAL_DIR}\nL√ºtfen zip'i doƒüru ≈üekilde a√ßtƒ±ƒüƒ±nƒ±zdan emin olun.")
    raise SystemExit

# Hyperparameters
BATCH_SIZE = 32
IMG_SIZE = 224
EPOCHS = 1  # quick verification
LR = 1e-3
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {DEVICE}")

# Transforms
train_transforms = transforms.Compose([
    transforms.RandomResizedCrop(IMG_SIZE),
    transforms.RandomHorizontalFlip(),
    transforms.ColorJitter(0.2,0.2,0.2,0.1),
    transforms.ToTensor(),
    transforms.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225])
])
val_transforms = transforms.Compose([
    transforms.Resize(IMG_SIZE+32),
    transforms.CenterCrop(IMG_SIZE),
    transforms.ToTensor(),
    transforms.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225])
])

# Datasets and loaders
train_dataset = datasets.ImageFolder(root=str(TRAIN_DIR), transform=train_transforms)
val_dataset = datasets.ImageFolder(root=str(VAL_DIR), transform=val_transforms)

# Use num_workers=0 on Windows/CPU for stability
train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True, num_workers=0, pin_memory=False)
val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=0, pin_memory=False)

classes = train_dataset.classes
num_classes = len(classes)
print("Classes:", classes)

# Model: prefer new weights API if available
try:
    weights = models.ResNet18_Weights.DEFAULT
    model = models.resnet18(weights=weights)
except Exception:
    # fallback to older API
    model = models.resnet18(pretrained=True)

for param in model.parameters():
    param.requires_grad = True
in_features = model.fc.in_features
model.fc = nn.Linear(in_features, num_classes)
model = model.to(DEVICE)

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=LR)

train_losses, val_losses = [], []
train_accs, val_accs = [], []

for epoch in range(1, EPOCHS+1):
    model.train()
    running_loss = 0.0
    running_corrects = 0
    total = 0

    for inputs, labels in train_loader:
        inputs = inputs.to(DEVICE)
        labels = labels.to(DEVICE)

        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item() * inputs.size(0)
        _, preds = torch.max(outputs, 1)
        running_corrects += torch.sum(preds == labels.data).item()
        total += inputs.size(0)

    epoch_loss = running_loss / max(total,1)
    epoch_acc = running_corrects / max(total,1)
    train_losses.append(epoch_loss)
    train_accs.append(epoch_acc)

    # validation
    model.eval()
    val_running_loss = 0.0
    val_running_corrects = 0
    val_total = 0
    with torch.no_grad():
        for inputs, labels in val_loader:
            inputs = inputs.to(DEVICE)
            labels = labels.to(DEVICE)
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            val_running_loss += loss.item() * inputs.size(0)
            _, preds = torch.max(outputs, 1)
            val_running_corrects += torch.sum(preds == labels.data).item()
            val_total += inputs.size(0)

    val_loss = val_running_loss / max(val_total, 1)
    val_acc = val_running_corrects / max(val_total, 1)
    val_losses.append(val_loss)
    val_accs.append(val_acc)

    print(f"Epoch {epoch}/{EPOCHS} ‚Äî train_loss: {epoch_loss:.4f}, train_acc: {epoch_acc:.4f}, val_loss: {val_loss:.4f}, val_acc: {val_acc:.4f}")

# Save model
os.makedirs("./models", exist_ok=True)
model_path = os.path.join("./models","animal_faces_resnet18_afhq.pth")
torch.save({'model_state_dict': model.state_dict(), 'classes': classes}, model_path)
print(f"Model kaydedildi: {model_path}")

# Plot
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.plot(train_losses, label='train_loss')
plt.plot(val_losses, label='val_loss')
plt.legend(); plt.title('Loss')
plt.subplot(1,2,2)
plt.plot(train_accs, label='train_acc')
plt.plot(val_accs, label='val_acc')
plt.legend(); plt.title('Accuracy')
plt.tight_layout()
plt.show()

# Quick inference on a few validation images
model.eval()
inputs, labels = next(iter(val_loader))
inputs = inputs.to(DEVICE)
labels = labels.to(DEVICE)
with torch.no_grad():
    n = min(8, inputs.size(0))
    outputs = model(inputs[:n])
    _, preds = torch.max(outputs, 1)

inv_norm = transforms.Normalize(
    mean=[-m/s for m,s in zip([0.485,0.456,0.406],[0.229,0.224,0.225])],
    std=[1/s for s in [0.229,0.224,0.225]]
)

fig, axes = plt.subplots(2,4, figsize=(12,6))
for i, ax in enumerate(axes.flatten()):
    if i < n:
        img = inv_norm(inputs[i].cpu())
        img = np.transpose(img.numpy(), (1,2,0))
        img = np.clip(img, 0, 1)
        ax.imshow(img)
        ax.set_title(f"Pred: {classes[preds[i]]}\nTrue: {classes[labels[i]]}")
    else:
        ax.axis('off')
    ax.axis('off')
plt.show()
```
