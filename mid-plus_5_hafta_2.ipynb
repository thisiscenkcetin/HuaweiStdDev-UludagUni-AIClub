```python
from datetime import datetime
import json

class Tool:
    def __init__(self, name, func, description):
        self.name = name
        self.func = func
        self.description = description

def get_current_time(query: str) -> str:
    """Åu anki saati dÃ¶ndÃ¼rÃ¼r."""
    return f"Åu anki saat: {datetime.now().strftime('%H:%M:%S')}"

def calculate_sum(numbers_str: str) -> str:
    """Verilen sayÄ±larÄ± toplar."""
    try:
        numbers = [float(x.strip()) for x in numbers_str.split(',')]
        result = sum(numbers)
        return f"{numbers_str} toplamÄ±: {result}"
    except ValueError:
        return "Hata: GeÃ§erli sayÄ±lar girin."

def multiply_numbers(numbers_str: str) -> str:
    """Verilen sayÄ±larÄ± Ã§arpar."""
    try:
        numbers = [float(x.strip()) for x in numbers_str.split(',')]
        result = 1
        for num in numbers:
            result *= num
        return f"{numbers_str} Ã§arpÄ±mÄ±: {result}"
    except ValueError:
        return "Hata: GeÃ§erli sayÄ±lar girin."

tools = [
    Tool(
        name="TimeGetter",
        func=get_current_time,
        description="Åu anki saati Ã¶ÄŸrenme iÅŸlemleri iÃ§in kullanÄ±lÄ±r."
    ),
    Tool(
        name="SumCalculator",
        func=calculate_sum,
        description="VirgÃ¼llerle ayrÄ±lmÄ±ÅŸ sayÄ±larÄ± toplamak iÃ§in kullanÄ±lÄ±r. Girdi: '2,3,5' ÅŸeklinde"
    ),
    Tool(
        name="MultiplyCalculator",
        func=multiply_numbers,
        description="VirgÃ¼llerle ayrÄ±lmÄ±ÅŸ sayÄ±larÄ± Ã§arpmak iÃ§in kullanÄ±lÄ±r. Girdi: '2,3,5' ÅŸeklinde"
    )
]

class Agent:
    def __init__(self, tools):
        self.tools = {tool.name: tool for tool in tools}
        self.tool_list = tools
    
    def get_tool_descriptions(self):
        descriptions = "KullanÄ±labilir araÃ§lar:\n"
        for tool in self.tool_list:
            descriptions += f"- {tool.name}: {tool.description}\n"
        return descriptions
    
    def run(self, query, expected_tool=None):
        print(f"\n> KullanÄ±cÄ± Sorgusu: {query}")
        print(f"\n> Agent DÃ¼ÅŸÃ¼nme SÃ¼reci:")
        print(f"  Sorgu analiz ediliyor...")
        
        if expected_tool:
            tool = self.tools[expected_tool]
            print(f"  '{expected_tool}' aracÄ± seÃ§ildi.")
        else:
            for tool_name, tool in self.tools.items():
                if any(keyword in query.lower() for keyword in ['topla', 'sum', 'ekle', '+']):
                    tool = self.tools['SumCalculator']
                    print(f"  'SumCalculator' aracÄ± seÃ§ildi.")
                    break
                elif any(keyword in query.lower() for keyword in ['Ã§arp', 'multiply', '*']):
                    tool = self.tools['MultiplyCalculator']
                    print(f"  'MultiplyCalculator' aracÄ± seÃ§ildi.")
                    break
                elif any(keyword in query.lower() for keyword in ['saat', 'time', 'zaman']):
                    tool = self.tools['TimeGetter']
                    print(f"  'TimeGetter' aracÄ± seÃ§ildi.")
                    break
            else:
                tool = self.tool_list[0]
                print(f"  '{tool.name}' aracÄ± seÃ§ildi.")
        
        print(f"\n> Tool Ã‡aÄŸrÄ±sÄ±:")
        print(f"  AraÃ§: {tool.name}")
        
        if 'Calculator' in tool.name:
            import re
            numbers = re.findall(r'\d+\.?\d*', query)
            numbers_str = ','.join(numbers)
            print(f"  ArgÃ¼manlar: numbers_str='{numbers_str}'")
            result = tool.func(numbers_str)
        else:
            print(f"  ArgÃ¼manlar: query='{query}'")
            result = tool.func(query)
        
        print(f"\n> Tool YanÄ±tÄ±:")
        print(f"  {result}")
        
        final_answer = f"Agent cevabÄ±: {result}"
        print(f"\n> Final Cevap:")
        print(f"  {final_answer}")
        
        return final_answer

agent = Agent(tools)

print("=" * 60)
print("AGENT VE TOOL SÄ°STEMÄ° - LangChain Demo")
print("=" * 60)

print("\n" + "=" * 60)
print("SORGU 1: Åu anki saat kaÃ§?")
print("=" * 60)
result1 = agent.run("Åu anki saat kaÃ§?")

print("\n" + "=" * 60)
print("SORGU 2: 15, 23, 42 sayÄ±larÄ±nÄ±n toplamÄ± kaÃ§tÄ±r?")
print("=" * 60)
result2 = agent.run("15, 23, 42 sayÄ±larÄ±nÄ±n toplamÄ± kaÃ§tÄ±r?")

print("\n" + "=" * 60)
print("SORGU 3: 5 ile 12 ile 3 sayÄ±larÄ±nÄ± Ã§arp")
print("=" * 60)
result3 = agent.run("5 ile 12 ile 3 sayÄ±larÄ±nÄ± Ã§arp")

print("\n" + "=" * 60)
print("SORGU 4: Saat kaÃ§ ve 10, 20, 30'u topla")
print("=" * 60)
result4 = agent.run("Saat kaÃ§ ve 10, 20, 30'u topla")
```
